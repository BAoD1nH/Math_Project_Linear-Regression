```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Đọc dữ liệu từ file CSV
data = pd.read_csv('train.csv')

# 1. Kiểm tra dữ liệu
print(data.head())        # Hiển thị những dòng đầu tiên của dữ liệu
print(data.info())        # Kiểm tra thông tin tổng quan về dữ liệu
print(data.describe())    # Thống kê mô tả
```

```python


# 2. Kiểm tra giá trị thiếu
print(data.isnull().sum())  # Đếm số lượng giá trị thiếu trong mỗi cột
```

```python

# 3. Phân tích mối quan hệ giữa các biến
# Ma trận tương quan
corr_matrix = data.corr()
print(corr_matrix)
```

```python

# Vẽ heatmap để trực quan hóa ma trận tương quan
plt.figure(figsize=(5, 4))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.show()
```

```python

# 4. Trực quan hóa dữ liệu
# Biểu đồ phân phối của Performance Index
sns.histplot(data['Performance Index'], kde=True)
plt.title('Distribution of Performance Index')
plt.show()
```

```python

# Biểu đồ phân tán giữa các yếu tố và Performance Index
sns.pairplot(data, y_vars='Performance Index', x_vars=['Hours Studied', 'Previous Scores', 'Extracurricular Activities', 'Sleep Hours', 'Sample Question Papers Practiced'], kind='scatter')
plt.show()
```
